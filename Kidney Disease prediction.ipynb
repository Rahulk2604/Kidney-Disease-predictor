```python
import pandas as pd
```

```python
Kd=pd.read_csv("D:\python\Project\kidney_disease.csv")
Kd
```

```python
Kd.info()
```

```python
Kd.describe()
```

```python
Kd = Kd.fillna(method='ffill')
Kd = Kd.fillna(method='bfill')
```

```python
Kd.isna().any()
```

```python
Kd=Kd.drop(['id'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

labelfit = le.fit(Kd[['rbc']])
Kd['rbc']=labelfit.transform(Kd[['rbc']])
Kd
```

```python

labelfit1 = le.fit(Kd[['pc']])
Kd['pc']=labelfit1.transform(Kd[['pc']])
Kd
```

```python
labelfit2 = le.fit(Kd[['pcc']])
Kd['pcc']=labelfit1.transform(Kd[['pcc']])
Kd
```

```python

labelfit1 = le.fit(Kd[['ba']])
Kd['ba']=labelfit1.transform(Kd[['ba']])
Kd
```

```python
le1=LabelEncoder()
for i in Kd.columns:
    Kd[i]=le1.fit_transform(Kd[i])
Kd
```

```python
x=Kd.iloc[:,0:24].values
y=Kd.iloc[:,-1].values
print(y)
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.5,random_state=0)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=20, criterion='gini',max_depth=8,max_features=8)
rfc.fit(xtrain,ytrain)
ypred = rfc.predict(xtest)
ypred
```

```python
training_score = rfc.score(xtrain,ytrain)
training_score
```

```python
from sklearn.metrics import accuracy_score,classification_report
RFCscore = accuracy_score(ytest , ypred)
RFCscore
```

```python

print("Classification Report",classification_report(ytest,ypred))
```

```python
from sklearn.preprocessing import StandardScaler    
sd= StandardScaler()    
xtrain= sd.fit_transform(xtrain)    
xtest= sd.transform(xtest)
```

```python
#K-nearest neighbour
from sklearn.neighbors import KNeighborsClassifier  
classifier= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
classifier.fit(xtrain, ytrain)
```

```python
Knnpred= classifier.predict(xtest)
Knnpred
```

```python
KNNscore = accuracy_score(ytest , Knnpred)
KNNscore
```

```python
#Naive bayes

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(xtrain,ytrain)
NBpred = nb.predict(xtest)
NBpred
```

```python
NBscore = accuracy_score(ytest , NBpred)
NBscore
```

```python
#SVM
from sklearn.svm import SVC
svc = SVC()
svc=svc.fit(xtrain,ytrain)
SVMpred = svc.predict(xtest)
SVMpred
```

```python
SVMscore = accuracy_score(ytest , SVMpred)
SVMscore
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain,ytrain)
LRpred = model.predict(xtest)
LRpred
```

```python
LRscore = accuracy_score(ytest , LRpred)
LRscore
```
